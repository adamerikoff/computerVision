```python
import torch
import numpy as np
import torch.nn as nn
```

```python
X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)
y = torch.tensor([[11.11], [22.22], [33.33], [44.44]], dtype=torch.float32)

print(X.shape)

X_test = torch.tensor([[6]], dtype=torch.float32)
```

```python
input_size = X.shape[1]
output_size = X.shape[1]
model = nn.Linear(input_size, output_size)

loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

print(f"Before training: f(6)={model(X_test)}")

for epoch in range(10000):
    optimizer.zero_grad()  # Zero gradients before backward pass

    y_hat = model(X)
    l = loss(y, y_hat)
    l.backward()  # Compute gradients
    optimizer.step()  # Update model parameters

    if epoch % 2 == 0:
        [w, b] = model.parameters()
        print(f"epoch {epoch+1}: w={w[0][0]:.3f}, loss={l.item():.8f}")
    if l < 0.00000001:
        break

print(f"After training: f(6)={model(X_test)}")
```
