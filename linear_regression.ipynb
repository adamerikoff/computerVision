```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate evenly spaced numpy array
x = np.linspace(0, 100, 50)
```

```python
# Create a straight line with some noise
slope = 2.0
intercept = 1.0
noise = np.random.normal(0, 20, 50)  
# Adding random noise
y = slope * x + intercept + noise
```

```python
# Create a scatter plot
plt.scatter(x, y, marker='o', c='b', label='Random Numbers')
plt.scatter(x.mean(), y.mean(), color='red', marker='o', label='Mean')
# Set axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Base Plot')

# Display a legend
plt.legend()

# Show the plot
plt.show()
```

```python
def y_predict(_x, _y):
    n = len(_x)
    _b1 = (n * np.sum(_x*_y) - np.sum(_x) * np.sum(_y)) / (n * np.sum(_x**2) - np.sum(_x) ** 2)
    _b0 = (np.sum(_y) - _b1 * np.sum(_x)) / n
    return _b1 * _x + _b0
```

```python
# Calculate y_predicted using the modified function
y_predicted = y_predict(x, y)

# Create a scatter plot
plt.scatter(x, y, marker='o', c='b', label='Random Numbers')
plt.plot(x, y_predicted, color='green', label='predicted')

# Set axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression')

# Display a legend
plt.legend()

# Show the plot
plt.show()
```
