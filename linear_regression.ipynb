```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate evenly spaced numpy array
x = np.linspace(0, 100, 50)
```

```python
# Create a straight line with some noise
slope = 2.0
intercept = 1.0
noise = np.random.normal(0, 20, 50)  
# Adding random noise
y = slope * x + intercept + noise
```

```python
# Calculation means
x_mean = np.sum(x) / x.shape[0]
y_mean = np.sum(y) / y.shape[0]
# Calculation xy
xy = x * y
# Calculation x^2
x_2 = np.power(x, 2)
x_2_sum = x_2.sum()
# Calculation sums
x_sum = x.sum()
y_sum = y.sum()
xy_sum = (x*y).sum()
```

```python
# Create a scatter plot
plt.scatter(x, y, marker='o', c='b', label='Random Numbers')
plt.scatter(x_mean, y_mean, color='red', marker='o', label='Mean')
# Set axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Base Plot')

# Display a legend
plt.legend()

# Show the plot
plt.show()
```

```python
variance = np.power(y-y_mean, 2)
std = np.sqrt(variance)
```

```python
b1 = (xy_sum - (x_sum * y_mean)) / ((x_mean * x_sum) - x_2_sum)
b0 = y_mean - (b1 * x_mean)
```

```python
def y_predict(x, b0, b1):
    return b0 + b1 * x
```

```python
y_predicted = y_predict(x, b0, b1)
```

```python
# Create a scatter plot
plt.scatter(x, y, marker='o', c='b', label='Random Numbers')
plt.plot(x, y_predicted, color='green', label='predicted')

# Set axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('no Linear regression')

# Display a legend
plt.legend()

# Show the plot
plt.show()
```
