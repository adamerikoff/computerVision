```python
# Import des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense

# Génération de données aléatoires pour les températures en Celsius (c) et Fahrenheit (f)
c = np.random.randint(-200, 200, 1000, dtype='int64')
f = ((c * 9) / 5) + 32

# Création d'un modèle de réseau de neurones séquentiel avec une couche dense
model = keras.Sequential()
model.add(Dense(units=1, input_shape=(1,), activation='linear'))

# Compilation du modèle avec la fonction de perte et l'optimiseur spécifiés
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(0.1))

# Entraînement du modèle sur les données de température Celsius et Fahrenheit
history = model.fit(c, f, epochs=500, verbose=0)
print("Entraînement terminé")

# Affichage de la prédiction pour 100 degrés Celsius
print(model.predict([100]))

# Affichage des poids (paramètres) du modèle
print(model.get_weights())

# Tracé de l'évolution de la perte (loss) au cours de l'entraînement
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```
