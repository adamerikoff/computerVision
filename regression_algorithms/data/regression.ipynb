```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./Experience-Salary.csv")

experience = data["exp(in months)"].values
salary = data["salary(in thousands)"].values
```

```python
# Adjust the figure size
plt.figure(figsize=(10, 8))  # Width x Height in inches

# Create a line plot
plt.scatter(experience, salary)

# Add labels and a title
plt.xlabel('experience')
plt.ylabel('salary')
plt.title('Keine Regression')

# Show the plot
plt.show()
```
