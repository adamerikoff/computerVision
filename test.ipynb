```python
import numpy as np
import matplotlib.pyplot as plt

# Generate evenly spaced numpy array
x = np.linspace(0, 100, 50)

# Create a straight line with some noise
slope = 2.0
intercept = 1.0
noise = np.random.normal(0, 20, 50)
y = slope * x + intercept + noise

# Calculation means
x_mean = np.sum(x) / x.shape[0]
y_mean = np.sum(y) / y.shape[0]

# Calculation xy
xy = x * y

# Calculation x^2
x_2 = np.power(x, 2)
x_2_sum = x_2.sum()

# Calculation sums
x_sum = x.sum()
y_sum = y.sum()
xy_sum = (x * y).sum()

# Calculate b1 and b0
b1 = (xy_sum - (x_sum * y_mean)) / ((x_sum * x_mean) - x_2_sum)
b0 = y_mean - b1 * x_mean

# Define the y_predict function
def y_predict(x, b0, b1):
    return b0 + -b1 * x

# Sort the x array in ascending order
x_sorted = np.sort(x)

# Create a scatter plot
plt.scatter(x, y, marker='o', c='b', label='Random Numbers')
plt.plot(x_sorted, y_predict(x_sorted, b0, b1), color='green', label='Linear Regression Line')

# Set axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

# Display a legend
plt.legend()

# Show the plot
plt.show()
```
