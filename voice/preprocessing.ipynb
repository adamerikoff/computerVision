```python
import librosa
import numpy as np

# Load the audio file
AUDIO_FILE = './data/voices/Julia_Gillard/24.wav'
samples, sample_rate = librosa.load(AUDIO_FILE, sr=None)

AUDIO_FILE2 = './data/voices/Benjamin_Netanyau/112.wav'
samples2, sample_rate2 = librosa.load(AUDIO_FILE2, sr=None)
```

```python
samples.size/sample_rate
```

```python
samples2.size/sample_rate2
```

```python
import librosa.display
import matplotlib.pyplot as plt

# Plot the waveform
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y=samples, sr=sample_rate, color="blue")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform AUDIO1')

# Plot the waveform
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y=samples2, sr=sample_rate2, color="blue")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform AUDIO2')
plt.show()
```

```python
# Print information about the audio data
print("AUDIO1")
print('Example shape:', samples.shape)
print('Sample rate:', sample_rate)
print('Data type:', type(samples))
print('First 40 samples:', samples[:40])
print(50*"=")
# Print information about the audio data2
print("AUDIO2")
print('Example shape:', samples2.shape)
print('Sample rate:', sample_rate2)
print('Data type:', type(samples2))
print('First 40 samples:', samples2[:40])
```

```python
sgram = librosa.stft(samples)
# Display the spectrogram
plt.figure(figsize=(8, 5))
librosa.display.specshow(sgram)

# Add labels to the plot
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Spectrogram')
plt.colorbar()

sgram2 = librosa.stft(samples2)
# Display the spectrogram
plt.figure(figsize=(8, 5))
librosa.display.specshow(sgram2)

# Add labels to the plot
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Spectrogram')
plt.colorbar()


plt.show()
```

```python
sgram_mag, _ = librosa.magphase(sgram)
mel_scale_sgram = librosa.feature.melspectrogram(S=sgram_mag, sr=sample_rate)
# Display the Mel spectrogram
plt.figure(figsize=(8, 5))
librosa.display.specshow(mel_scale_sgram, sr=sample_rate, x_axis='time', y_axis='mel')

# Add labels to the plot
plt.xlabel('Time')
plt.ylabel('Mel Frequency')
plt.title('Mel Spectrogram')
plt.colorbar()

sgram_mag2, _ = librosa.magphase(sgram2)
mel_scale_sgram2 = librosa.feature.melspectrogram(S=sgram_mag2, sr=sample_rate2)
# Display the Mel spectrogram
plt.figure(figsize=(8, 5))
librosa.display.specshow(mel_scale_sgram2, sr=sample_rate2, x_axis='time', y_axis='mel')

# Add labels to the plot
plt.xlabel('Time')
plt.ylabel('Mel Frequency')
plt.title('Mel Spectrogram')
plt.colorbar()

plt.show()

```

```python
plt.figure(figsize=(8, 5))
mel_sgram = librosa.amplitude_to_db(mel_scale_sgram, ref=np.min)
librosa.display.specshow(mel_sgram, sr=sample_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time (s)')
plt.ylabel('Mel Frequency (Hz)')
plt.title('Mel Spectrogram')

plt.figure(figsize=(8, 5))
mel_sgram2 = librosa.amplitude_to_db(mel_scale_sgram2, ref=np.min)
librosa.display.specshow(mel_sgram2, sr=sample_rate2, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time (s)')
plt.ylabel('Mel Frequency (Hz)')
plt.title('Mel Spectrogram')

plt.show()
```
