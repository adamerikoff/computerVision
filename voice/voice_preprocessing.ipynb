```python
import numpy as np
n_fft = 2048
ft = np.abs(librosa.stft(y[:n_fft], hop_length = n_fft+1))
plt.plot(ft)
plt.title('Spectrum')
plt.xlabel('Frequency Bin')
plt.ylabel('Amplitude')
```

```python
spec = np.abs(librosa.stft(y, hop_length=512))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
```

```python
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time')
plt.title('Mel Spectrogram')
plt.colorbar(format='%+2.0f dB')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the sampling rate (number of samples per second)
sampling_rate = 1000  # You can adjust this value to change the resolution of the plot

# Set the time duration for the plot
duration = 1  # 1 second

# Generate time values for the duration specified
time = np.linspace(0, duration, duration * sampling_rate)

# Set the frequency of the sine wave
frequency = 40  # 40 Hz

# Calculate the amplitude of the sine wave
amplitude = 1

# Generate the sine wave using the formula y(t) = A * sin(2 * pi * f * t)
sine_wave = amplitude * np.sin(2 * np.pi * frequency * time)

# Plot the sine wave
plt.plot(time, sine_wave)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sine Wave (40 Hz)')
plt.grid(True)
plt.show()
```
